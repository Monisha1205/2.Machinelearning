```python
import pandas as pd # pandas used for working with data sets
```

```python
import numpy as np  #NumPy can be used to perform a wide variety of mathematical operations on arrays
```

```python
import matplotlib.pyplot as plt # matplotlib used for plot graph
```

```python
dataset=pd.read_csv("Salary_Data.csv")# assign variable has dataset by read data file csv
dataset
```

```python
# split input and output variables
independed=dataset[["YearsExperience"]] # assign input variable 
independed
```

```python
depended=dataset[["Salary"]]# assign output variable
depended
```

```python
# plot two variable x, y axis
# plot x label font size
# plot y label font size
# plot graph show using this we can ideatify model is linear regression
plt.scatter(independed,depended)
plt.xlabel('YearsExperience',fontsize=20)
plt.ylabel('salary',fontsize=20)
plt.show()
```

```python
# split train and test set using sklearn model import libreary 
from sklearn.model_selection import train_test_split
# assign x train, test ,y train, test in random selection
x_train,x_test,y_train,y_test=train_test_split(independed,depended,test_size=0.30,random_state=0)
```

```python
# linear regression model is fit using train dataset
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_ # coefficient of input is find
weight
```

```python
bias=regressor.intercept_ # intercept value is find
bias
```

```python
# our predicted value is find using test set it is Y_predict becaused this valuse is need to find R2
y_pred=regressor.predict(x_test) 
y_pred
```

```python
# validating parametres R2 finding using y test, y pred
from sklearn.metrics import r2_score 
r_score=r2_score(y_test,y_pred)
r_score
```

```python
# pickle library used for store in new file 
import pickle 
filename="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,"wb")) # create new file
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav","rb")) # load model for recall it in deloyment stage 
result=loaded_model.predict([[15],[0]])# test user prediction 
result # final output 
```
